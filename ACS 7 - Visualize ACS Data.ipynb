```python
def define_weighted_symbology(parcel_layer, weighted_color_ramp, classification_field, break_count, weighted_symbol_outline_color, weighted_symbol_outline_width): 
    
    # Identify map 
    aprx = arcpy.mp.ArcGISProject("CURRENT")
    
    # Initialize weighted_symbology 
    weighted_symbology = parcel_layer.symbology 
    
    # Define weighted Colors Renderer
    weighted_symbology.updateRenderer("GraduatedColorsRenderer") 

    # Define field to weighted_symbology.bolize off of and how many levels to have 
    weighted_symbology.renderer.classificationField = classification_field
    weighted_symbology.renderer.classificationMethod = "NaturalBreaks"
    weighted_symbology.renderer.breakCount = break_count

    # Define color scheme - this assumes a basic color ramp 
    weighted_symbology.renderer.colorRamp = aprx.listColorRamps(weighted_color_ramp)[0]
    
    for cb in weighted_symbology.renderer.classBreaks: 
        cb.symbol.outlineColor = weighted_symbol_outline_color
        cb.symbol.outlineWidth = weighted_symbol_outline_width
            
    # Apply weighted_symbology
    parcel_layer.symbology = weighted_symbology 
    
    return parcel_layer.symbology
```
