```python
def find_unique_values(focus_features, focus_field): 
    
    unique_values = [] 
    
    with arcpy.da.SearchCursor(focus_features, [focus_field]) as cursor: 
        for row in cursor: 
            if row[0] not in unique_values: 
                unique_values.append(row[0])

    return 
```
